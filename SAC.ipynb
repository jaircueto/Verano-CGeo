# SAC

```julia
using SACFiles
using PyPlot
```

```julia
sacE=readsac("SV.ST03L..EHE.SAC");
sacN=readsac("SV.ST03L..EHN.SAC");
sacZ=readsac("SV.ST03L..EHZ.SAC");
```

```julia
datE = sacE.data;
datN = sacN.data;
datZ = sacZ.data;
```

```julia
plot(datE);
```

```julia
plot(datN);
```

```julia
plot(datZ);
```

```julia
tsE=(collect(0:1:length(datE)-1))*sacE.hdr.delta;
tsN=(collect(0:1:length(datN)-1))*sacN.hdr.delta;
tsZ=(collect(0:1:length(datZ)-1))*sacZ.hdr.delta;
```

```julia
plot(tsE,datE);
```

```julia
plot(tsN,datN);
```

```julia
plot(tsZ,datZ);
```

```julia
plot(FtraN);
```

```julia
plot(FtraZ);
```

```julia
fxE=collect((0:1:9000-1)*(100/length(FtraE)));
fxN=collect((0:1:9000-1)*(100/length(FtraN)));
fxZ=collect((0:1:9000-1)*(100/length(FtraZ)));
```
