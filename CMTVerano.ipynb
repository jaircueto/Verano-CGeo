```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=12,8
```

```python
import obspy
```

```python
CMT = obspy.read_events('CMTMex.txt')
```

```python
CMT.plot(projection='local',method='basemap',title=False);
```

```python
CMTMex = CMT.filter("latitude > 11.9", "latitude < 32.9","longitude > -104", "longitude < -85")
```

```python
CMTMex.plot(projection='local',method='basemap',title=False);
```
